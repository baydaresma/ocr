```python
from pdf2image import convert_from_path
import easyocr
import os

# 📂 3. Google Drive'dan dosya almak istiyorsan:
# from google.colab import drive
# drive.mount('/content/drive')
# pdf_path = "/content/drive/MyDrive/ornek.pdf"

# 👇 PDF dosyanı Colab’e sürükle-bırak yaptıysan:
pdf_path = "/content/disiplin_amirleri_yonetmeligi.pdf"  # kendi PDF dosya adını buraya yaz

# 🖼️ 4. PDF’i sayfa sayfa resimlere çevir
pages = convert_from_path(pdf_path, dpi=300)

# 📖 5. EasyOCR okuyucu (Türkçe)
reader = easyocr.Reader(['tr'])

# 🧾 6. Tüm sayfalardan metin çıkar
all_text = ""
for i, page in enumerate(pages):
    image_path = f"page_{i+1}.jpg"
    page.save(image_path, 'JPEG')

    print(f"🔍 {i+1}. sayfa okunuyor...")
    result = reader.readtext(image_path, detail=0)
    all_text += "\n".join(result) + "\n\n"

# 💾 7. Çıkan metni dosyaya kaydet
with open("metin_cikti.txt", "w", encoding="utf-8") as f:
    f.write(all_text)

print("✅ OCR tamamlandı! metin_cikti.txt dosyası oluşturuldu.")
```

```python
# Gerekli kütüphaneler
!pip install pytesseract pdf2image opencv-python-headless pillow
!apt-get install -y tesseract-ocr tesseract-ocr-tur poppler-utils

import cv2
import pytesseract
import numpy as np
from pdf2image import convert_from_path
from pathlib import Path
from PIL import Image

# PDF yolu
pdf_path = "/content/disiplin_amirleri_yonetmeligi.pdf"

# Çıktı klasörü
output_dir = Path("ocr_output")
output_dir.mkdir(exist_ok=True)

# PDF'i sayfa sayfa resme dönüştür
pages = convert_from_path(pdf_path)

def preprocess_image(img):
    """OCR için görseli iyileştir"""
    gray = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return thresh

def detect_table_cells(page_image):
    """OpenCV ile tablo hücrelerini tespit et"""
    img = preprocess_image(page_image)

    kernel_h = cv2.getStructuringElement(cv2.MORPH_RECT, (40,1))
    kernel_v = cv2.getStructuringElement(cv2.MORPH_RECT, (1,40))

    detect_h = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_h, iterations=2)
    detect_v = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_v, iterations=2)

    grid = cv2.add(detect_h, detect_v)
    contours, _ = cv2.findContours(grid, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    cells = []
    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if w > 20 and h > 20:  # çok küçük alanları filtrele
            cell_img = img[y:y+h, x:x+w]
            cells.append((x, y, w, h, cell_img))

    # Yukarıdan aşağıya, soldan sağa sırala
    cells = sorted(cells, key=lambda b: (b[1], b[0]))
    return cells

def ocr_cells(cells):
    """Tablo hücrelerini OCR ile oku"""
    texts = []
    for x, y, w, h, cell_img in cells:
        text = pytesseract.image_to_string(cell_img, lang='tur', config="--psm 6").strip()
        if text:
            texts.append(text)
    return texts

def ocr_page(page_image):
    """Sayfadaki düz metni OCR ile oku"""
    img = preprocess_image(page_image)
    text = pytesseract.image_to_string(img, lang='tur', config="--psm 1").strip()
    return text

# Tüm sayfalar için OCR
all_texts = []

for i, page in enumerate(pages):
    print(f"Sayfa {i+1} işleniyor...")

    # Tablo hücreleri
    cells = detect_table_cells(page)
    cell_texts = ocr_cells(cells)

    # Sayfa düz metni
    page_text = ocr_page(page)

    # Birleştir
    all_texts.append(page_text)
    all_texts.extend(cell_texts)

# TXT olarak kaydet
txt_path = output_dir / "output.txt"
with open(txt_path, "w", encoding="utf-8") as f:
    f.write("\n".join(all_texts))

print(f"OCR tamamlandı ✅ Metinler {txt_path} içinde.")
```

```python
!pip install pytesseract pdf2image pillow opencv-python-headless
!apt-get install -y tesseract-ocr tesseract-ocr-tur poppler-utils

import pytesseract
from pdf2image import convert_from_path
from pathlib import Path
import cv2
import numpy as np

pdf_path = "/content/disiplin_amirleri_yonetmeligi.pdf"
output_dir = Path("ocr_output")
output_dir.mkdir(exist_ok=True)

pages = convert_from_path(pdf_path)

def preprocess_image(img):
    img_np = np.array(img)
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return thresh

def detect_table_cells(page_img):
    img = preprocess_image(page_img)
    # Satır ve sütun çizgilerini bul
    kernel_h = cv2.getStructuringElement(cv2.MORPH_RECT, (50,1))
    kernel_v = cv2.getStructuringElement(cv2.MORPH_RECT, (1,50))
    detect_h = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_h, iterations=2)
    detect_v = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_v, iterations=2)
    grid = cv2.add(detect_h, detect_v)
    contours, _ = cv2.findContours(grid, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    cells = []
    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if w > 20 and h > 20:
            cell_img = img[y:y+h, x:x+w]
            cells.append((x, y, w, h, cell_img))
    # Yukarıdan aşağıya, soldan sağa sırala
    cells = sorted(cells, key=lambda b: (b[1], b[0]))
    return cells

all_texts = []

for i, page in enumerate(pages):
    print(f"Sayfa {i+1} işleniyor...")
    # Hücre OCR
    cells = detect_table_cells(page)
    cell_texts = []
    for x, y, w, h, cell_img in cells:
        # Hücreyi biraz büyüt (margin)
        margin = 3
        #crop = cell_img[max(0,y-margin):y+h+margin, max(0,x-margin):x+w+margin]
        # Orijinal sayfa resmi
        page_np = np.array(page)

        x1 = max(0, x-margin)
        y1 = max(0, y-margin)
        x2 = min(page_np.shape[1], x+w+margin)
        y2 = min(page_np.shape[0], y+h+margin)

        crop = page_np[y1:y2, x1:x2]

        text = pytesseract.image_to_string(crop, lang='tur', config="--psm 6").strip()
        if text:
            cell_texts.append(text)
    all_texts.extend(cell_texts)

# TXT olarak kaydet
txt_path = output_dir / "output_cells.txt"
with open(txt_path, "w", encoding="utf-8") as f:
    f.write("\n".join(all_texts))

print(f"Hücre OCR tamamlandı ✅ {txt_path} içinde")
```
